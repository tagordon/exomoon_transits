```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
from matplotlib.colors import colorConverter
```

```python
def draw_geometry(axs, rp, rm, xp, yp, xm, ym):
    
    star = plt.Circle((0, 0), 1, color='#ffae36', alpha=-0.75)
    star_outline = plt.Circle((0, 0), 1, color=colorConverter.to_rgba('w', alpha=0.0), ec='k')
    
    planet = plt.Circle((xp, yp), rp, color=colorConverter.to_rgba('r', alpha=0.5), ec='k')
    planet_base = plt.Circle((xp, yp), rp, color='w')
    moon = plt.Circle((xm, ym), rm, color=colorConverter.to_rgba('b', alpha=0.5), ec='k')
    moon_base = plt.Circle((xm, ym), rm, color='w')
    
    axs.add_patch(star)
    axs.add_patch(planet_base)
    axs.add_patch(moon_base)
    axs.add_patch(star_outline)
    
    axs.add_patch(planet)
    axs.add_patch(moon)
```

```python
plt.figure(figsize=(10, 10))
plt.ylim(-1.5, 1.5)
plt.xlim(-1.5, 1.5)
ax = plt.gca()

draw_geometry(ax, 0.45, 0.2, 0.7, 0.0, 0.3, -0.6)
plt.plot(0, 0, 'ko')
plt.plot(0.7, 0.0, 'ko')
plt.plot(0.3, -0.6, 'ko')

plt.plot((0.003, 0.7), (0.0, 0.0), 'k')
plt.plot((0.3, 0.7), (-0.61, 0.0), 'k')

angle_arc = Arc((0.7, 0.0), 0.15, 0.15, theta1 = 180, theta2 = 230)
ax.add_patch(angle_arc)

ax.annotate(r"$b_p$", (0.1, 0.05), fontsize=15)
ax.annotate(r"$b_{pm}$", (0.6, -0.3), fontsize=15)
ax.annotate(r"$\theta$", (0.55, -0.1), fontsize=15)

ax.axis('off')
plt.savefig("/Users/tgordon/Desktop/coords.pdf", bbox_inches='tight')
```
