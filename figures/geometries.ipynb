```python
import matplotlib.pyplot as plt
from matplotlib.colors import colorConverter
```

```python
def draw_geometry(axs, rp, rm, xp, yp, xm, ym):
    
    star = plt.Circle((0, 0), 1, color='#ffae36', alpha=0.75)
    star_outline = plt.Circle((0, 0), 1, color=colorConverter.to_rgba('w', alpha=0.0), ec='k')
    
    planet = plt.Circle((xp, yp), rp, color=colorConverter.to_rgba('r', alpha=0.5), ec='k')
    planet_base = plt.Circle((xp, yp), rp, color='w')
    moon = plt.Circle((xm, ym), rm, color=colorConverter.to_rgba('b', alpha=0.5), ec='k')
    moon_base = plt.Circle((xm, ym), rm, color='w')
    
    axs.add_patch(star)
    axs.add_patch(planet_base)
    axs.add_patch(moon_base)
    axs.add_patch(star_outline)
    
    axs.add_patch(planet)
    axs.add_patch(moon)
```

```python
fig, axs = plt.subplots(4, 4, sharex=True, sharey=True, figsize=(25, 25))
[ax.set_ylim(-0.3, 1.5) for ax in axs.flatten()];
[ax.set_xlim(-0.3, 1.5) for ax in axs.flatten()];
[ax.axes.xaxis.set_visible(False)for ax in axs.flatten()];
[ax.axes.yaxis.set_visible(False)for ax in axs.flatten()];
plt.subplots_adjust(hspace=0.1, wspace=0.1)

rp, rm = 0.4, 0.3

draw_geometry(axs[0, 0], rp, rm, 1.0, 1.08, 1.15, 0.75)
axs[0, 0].annotate("A", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[0, 1], rp, rm, 1.0, 1.08, 0.7, 0.75)
axs[0, 1].annotate("B", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[0, 2], rp, rm, 0.8, 0.8, 1.15, 0.75)
axs[0, 2].annotate("C", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[0, 3], rp, rm, 0.3, 0.3, 0.3, 0.3)
axs[0, 3].annotate("D", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)

draw_geometry(axs[1, 0], rp, rm, 0.3, 0.3, 0.5, 0.1)
axs[1, 0].annotate("E", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[1, 1], rp, rm, 0.3, 0.3, 0.65, 0.65)
axs[1, 1].annotate("F", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[1, 2], rp, rm - 0.05, 0.5, 0.5, 0.42, 0.42)
axs[1, 2].annotate("G", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[1, 3], rp, rm, 0.6, 0.6, 0.3, 0.3)
axs[1, 3].annotate("H", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)

draw_geometry(axs[2, 0], rp, rm, 0.65, 0.65, 0.65, 0.65)
axs[2, 0].annotate("I", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[2, 1], rp, rm, 0.6, 0.6, 0.75, 0.75)
axs[2, 1].annotate("J", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[2, 2], rp, rm, 0.8, 0.8, 0.65, 0.65)
axs[2, 2].annotate("K", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[2, 3], rp, rm + 0.03, 0.45, 0.45, 0.7, 0.7)
axs[2, 3].annotate("L", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)

draw_geometry(axs[3, 0], rp, rm + 0.03, 0.9, 0.9, 0.7, 0.7)
axs[3, 0].annotate("M", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[3, 1], rp, rm, 0.8, 1.07, 1.1, 0.5)
axs[3, 1].annotate("N", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[3, 2], rp, rm, 0.4, 0.67, 0.73, 0.18)
axs[3, 2].annotate("O", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)
draw_geometry(axs[3, 3], rp, rm, 0.6, 0.65, 0.9, 0.4)
axs[3, 3].annotate("P", xy=(0.05, 0.9), xycoords='axes fraction', fontsize=30)

plt.tight_layout()
plt.savefig('/Users/tgordon/Desktop/configurations.pdf')
```
