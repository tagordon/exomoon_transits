```python
#dt = np.diff(t)[0]
import time
import rebound

sim = rebound.Simulation()
sim.units = ('d', 'AU', 'Mearth')
sim.add(m=ms)
sim.add(m=mp, e=ep, Omega=Op + np.pi, P=Pp, inc=ip, omega=wp, T=t0p)
sim.add(m=mm, e=em, Omega=Om + np.pi, P=Pm, inc=im, omega=wm, T=t0m, primary=sim.particles[1])
sim.integrator = "WHfast"
sim.dt = 1e-3

xp_r, yp_r, zp_r = np.zeros(len(t)), np.zeros(len(t)), np.zeros(len(t))
xm_r, ym_r, zm_r = np.zeros(len(t)), np.zeros(len(t)), np.zeros(len(t))
start_time = time.time()
for i, ti in enumerate(t):
    sim.integrate(ti, exact_finish_time=0)
    xp_r[i] = sim.particles[1].x
    yp_r[i] = sim.particles[1].y
    zp_r[i] = sim.particles[1].z
    xm_r[i] = sim.particles[2].x
    ym_r[i] = sim.particles[2].y
    zm_r[i] = sim.particles[2].z
print(time.time() - start_time)
```
